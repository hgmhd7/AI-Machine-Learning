##LeetCode DataSet
It contains the following info:

- **id**: problem id
- **title**: problem name
- **description**: problem description
- **is_premium**: whether the questions requires a premium account
- **difficulty**: easy, medium, or hard
- **solution_link**: how often the answer submitted is correct
- **acceptance_rate**: how often the answer submitted is correct
- **frequency**: how often the problem is attempted
- **url**: url to the problem
- **discuss_count**: how many comments are submitted by users
- **accepted**: how many times the answer was accepted
- **submissions**: how many times the answer was submitted
- **companies**: which companies were tagged as having asked this specific problem
- **related_topics**: related topics to the current problem
- **likes**: how many likes the problem got
- **dislikes**: how many dislikes the problem got
- **rating**: likes / (likes + dislikes)
- **asked_by_faang**: whether or not the question was asked by facebook, apple, amazon, google, or netflix
- **similar_questions**: similar problems with problem name, slug, and difficulty

### Question 6

#### Handle the missing values in the columns

```python
#Solution 6(your code here)

# Check the number of misssing values in column
leetcode_df['solution_link'].isna().sum()
```

```python
"""Delete the solution_link column since about hald of them are missing.
I dont really see a reasonable way to imputate the missng values since such
a huge portion of that column is missing.  more importantly, every value in the column
is uniqie, to there is no real mode..."""

leetcode_df.drop('solution_link', axis=1, inplace=True)
```

```python
"""Fill the NA values in the 'company' column with Google since it
is the mode.  Furthermore, there is a clear mode here with Google
and there are very few missing values in this column relative to
the total number of rows."""

leetcode_df['companies'] = leetcode_df['companies'].fillna('Google')
```

```python
"""Fill the NA values in the 'related_topics' column with Array since it
is the mode.  There are more missing values in this column, bu I think
it is within reason to still replace the missign values with the mode."""

leetcode_df['related_topics'] = leetcode_df['related_topics'].fillna('Array')
```

```python
"""Since more than half the values in the 'similar_questions' are missing,
it makes sense to just to drop the column."""

leetcode_df.drop('similar_questions', axis=1, inplace=True)
```

### Question 7

#### Convert the datatypes for **accepted** and **submissions** columns

```python
#Solution 7(your code here)
# Convert 'accepted' and 'submissions' to numeric types

""" Strip the last character from the accepted and submissions column.  If the column
ends with M, multiply it by 1,000,000 to convert to millions and 1,000 to convert to
thousands.  after this """


# Get the filtered indicies for the columns ending in M and K
M_filtered_accepted_indices = leetcode_df['accepted'].str.get(-1) == 'M'
K_filtered_accepted_indices = leetcode_df['accepted'].str.get(-1) == 'K'

M_filtered_submissions_indices = leetcode_df['submissions'].str.get(-1) == 'M'
K_filtered_submissions_indices = leetcode_df['submissions'].str.get(-1) == 'K'


# Stip the M and K from the columns ending in those characters,
# convert to numbers, and multiply by the appropriate multiplier...
leetcode_df.loc[M_filtered_accepted_indices, 'accepted'] = pd.to_numeric(leetcode_df.loc[M_filtered_accepted_indices, 'accepted'].str.rstrip('M')) * 1000000
leetcode_df.loc[K_filtered_accepted_indices, 'accepted'] = pd.to_numeric(leetcode_df.loc[K_filtered_accepted_indices, 'accepted'].str.rstrip('K')) * 1000

leetcode_df.loc[M_filtered_submissions_indices, 'submissions'] = pd.to_numeric(leetcode_df.loc[M_filtered_submissions_indices, 'submissions'].str.rstrip('M')) * 1000000
leetcode_df.loc[K_filtered_submissions_indices, 'submissions'] = pd.to_numeric(leetcode_df.loc[K_filtered_submissions_indices, 'submissions'].str.rstrip('K')) * 1000





```

```python
# Describe numerical columns to check the min and max for like and dislikes before I remove outliers below
leetcode_df.describe(percentiles=[0.05, 0.3, 0.95])
```

```python
# Calculating IQR for 'liikes'
likes_Q1 = leetcode_df['likes'].quantile(0.25)
likes_Q3 = leetcode_df['likes'].quantile(0.75)
likes_IQR = likes_Q3 - likes_Q1
likes_lower_bound = likes_Q1 - 1.5 * likes_IQR
likes_upper_bound = likes_Q3 + 1.5 * likes_IQR


# Calculating IQR for 'Dislikes'
dislikes_Q1 = leetcode_df['dislikes'].quantile(0.25)
dislikes_Q3 = leetcode_df['dislikes'].quantile(0.75)
dislikes_IQR = dislikes_Q3 - dislikes_Q1
dislikes_lower_bound = dislikes_Q1 - 1.5 * dislikes_IQR
dislikes_upper_bound = dislikes_Q3 + 1.5 * dislikes_IQR


# Treating likes outliers by capping
leetcode_df['likes'] = leetcode_df['likes'].apply(lambda x: likes_upper_bound if x > likes_upper_bound else (likes_lower_bound if x < likes_lower_bound else x))

# Treating dislikes outliers by capping
leetcode_df['dislikes'] = leetcode_df['dislikes'].apply(lambda x: dislikes_upper_bound if x > dislikes_upper_bound else (dislikes_lower_bound if x < dislikes_lower_bound else x))
```
